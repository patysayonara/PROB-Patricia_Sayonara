Por meio do uso do dataset "basedados.xlsx", o qual contém dados relacionados à Mega-Sena, objetiva-se analisar a relação entre os números mais sorteados e os ganhadores, tanto da Mega-Sena, quanto da Quina e da Quadra.

```python
import pandas as pd
arquivo = 'basedados.xlsx'
excel = pd.ExcelFile(arquivo)
data = excel.parse(0)
data
```

```python
#Como apresentado abaixo, não há presença de dados faltantes neste dataset, logo, não foi necessário tratamento dos dados
data.isnull().sum()
```

```python
#Imagem 1
import matplotlib.pyplot as plt
fig = plt.figure(figsize =(20, 10))
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
ax1.plot(data['Data Sorteio'][1:], data['Ganhadores_Sena'][1:])
ax1.set_title("Frequência de Ganhadores da Mega-Sena")
ax2.plot(data['Data Sorteio'][1:], data['Ganhadores_Quina'][1:])
ax2.set_title("Frequência de Ganhadores da Quina")
ax3.plot(data['Data Sorteio'][1:], data['Ganhadores_Quadra'][1:])
ax3.set_title("Frequência de Ganhadores da Quadra")
plt.show()
```

A imagem acima apresenta a frequência de ganhadores da Mega-Sena, da Quina e da Quadra ao longo de todo o período disponível no dataset. Contudo, pode-se perceber um crescimento nas frequências por volta de, aproximadamente, 2010 em diante. Logo, o período a partir desses ano foi escolhido como objeto específico desta análise.

```python
#Imagem 2
fig, ax = plt.subplots()
num_cols = ['Dezena1', 'Dezena2', 'Dezena3', 'Dezena4', 'Dezena5', 'Dezena6']
ax.boxplot(data[num_cols][data['Data Sorteio'] >= '2010-01-01'].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylabel("Números que podem ser sorteados")
ax.set_ylim(0, 60)
plt.show()
```

A imagem acima apresenta a distribuição dos números sorteados em cada dezena dentre os possíveis, entre 0 e 60. Pode-se perceber que há uma grande abrangência em todos e que, também, pelo menos 50% dos números sorteados em cada dezena concentra-se entre 16 e 45, aproximadamente. Tal fato impulsionou a decisão de analisar em seguida qual a relação entre os números com maior frequência de sorteio e os números selecionados dos sorteios com mais vencedores.

```python
#15 números mais sorteados a partir de 2010
dezenas = ["Dezena1", "Dezena2", "Dezena3", "Dezena4", "Dezena5", "Dezena6"]
data[data["Data Sorteio"] >= "2010-01-01"][dezenas].unstack().value_counts()[0:15]
```

Diante dos números apresentados acima, pode-se perceber que entre os 15 números sorteados, 10 estão entre os números 16 e 45, o que se correlaciona com os dados apresentados na imagem 2.

```python
#À esquerda estão as linhas em que se encontram os dados relacionados aos 5 sorteios em que se tiveram a maior quantidade de 
# vencedores da Mega-Sena
data[data['Data Sorteio'] >= '2010-01-01']['Ganhadores_Sena'].sort_values(ascending=False)[0:5]
```

```python
#Visualização das linhas encontradas anteriormente
num1 = [1999, 1210, 1889, 1774, 1349]
data.loc[num1]
```

```python
#Aplicação do mesmo método anterior, agora para as 5 maiores frequências de ganhadores da Quina
data[data['Data Sorteio'] >= '2010-01-01']['Ganhadores_Quina'].sort_values(ascending=False)[0:5]
```

```python
#Visualização dos dados
num2 = [1999, 1664, 1889, 1244, 1454]
data.loc[num2]
```

```python
#Aplicação do método de achar as maiores frequências agora para achar os 5 sorteios com mais ganhadores da Quadra
data[data['Data Sorteio'] >= '2010-01-01']['Ganhadores_Quadra'].sort_values(ascending=False)[0:5]
```

```python
#Visualização dos dados
num3 = [1999, 1664, 1889, 1454, 1244]
data.loc[num3]
```

Por meio dos dados relacionados às maiores frequências de ganhadores da Mega-Sena, Quina e Quadra, juntamente com os números mais sorteados a partir de 2010, pode-se perceber que há pelo menos um dos 15 números mais sorteados entre as dezenas sorteadas nos concursos com maior frequência de ganhadores. Logo, pode-se estabelecer uma forte possibilidade de presença de no mínimo 1 desses números no sorteio, assim como maior probabilidade de escolha desses números pelos apostadores.

```python
#Imagem 3
fig, ax = plt.subplots()
ax.scatter(data['Ganhadores_Quina'][data['Data Sorteio'] >= '2010-01-01'].sort_values(ascending=False), data['Ganhadores_Quadra'][data['Data Sorteio'] >= '2010-01-01'].sort_values(ascending=False))
ax.set_xlabel("Quina")
ax.set_ylabel("Quadra")
ax.set_xlim(0,1500)
plt.show()
```

Devido aos dados apresentados nos sorteios de Quina e Quadra com maior frequência de ganhadores, pode-se perceber que houve uma interseção entre os sorteios, ou seja, foram achados os mesmos sorteios para ambas as análises. Isso tornou também possível a produção do gráfico apresentado na imagem acima, no qual percebe-se uma forte correlação entre os sorteios de maior frequência de ganhadores da Quina e da Quadra.
