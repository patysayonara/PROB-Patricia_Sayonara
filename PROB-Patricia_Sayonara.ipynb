```python
import pandas as pd
arquivo = 'basedados.xlsx'
excel = pd.ExcelFile(arquivo)
data = excel.parse(0)
data
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize =(20, 10))
ax1 = fig.add_subplot(3, 1, 1)
ax2 = fig.add_subplot(3, 1, 2)
ax3 = fig.add_subplot(3, 1, 3)
ax1.plot(data['Data Sorteio'][1:], data['Ganhadores_Sena'][1:])
ax2.plot(data['Data Sorteio'][1:], data['Ganhadores_Quina'][1:])
ax3.plot(data['Data Sorteio'][1:], data['Ganhadores_Quadra'][1:])
plt.show()
```

```python
#Percebe-se um crescimento na frequência, aproximadamente, por volta de 2010 em diante, logo, foi escolhido os dados provenientes
# de datas a partir de 2010 como objetos de estudo
```

```python
fig, ax = plt.subplots()
num_cols = ['Dezena1', 'Dezena2', 'Dezena3', 'Dezena4', 'Dezena5', 'Dezena6']
ax.boxplot(data[num_cols][data['Data Sorteio'] >= '2010-01-01'].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylim(0, 60)
plt.show()
```

```python
#Percebe-se uma concentração de pelo menos 50% em cada dezena entre os números 16 e 45, aproximadamente
```

```python
#Objetiva-se analisar a relação entre os números selecionados dos sorteios com mais vencedores e os números com maior frequência
# de sorteio
```
